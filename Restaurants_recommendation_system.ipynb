Preprocessing techniques:

The preprocessing techniques applied in this code ensure that the dataset is clean, consistent, and ready for further analysis. These techniques include:

1️⃣ Handling Missing Values in restaurant_name: Since the restaurant name is a crucial identifier, any row where restaurant_name is missing is removed. This prevents incomplete or ambiguous entries from affecting the dataset.

2️⃣ Dropping Irrelevant Columns (additional_image_url): If the column additional_image_url exists, it is removed, as it is not essential for data analysis or recommendation system processing. Removing unnecessary columns optimizes storage and computation efficiency.

3️⃣ Filling Missing Categories: If the category column has missing values, they are replaced with the Arabic term "مطعم" (meaning "restaurant"). This ensures consistency in classification and avoids unclassified entries. Arabic text reshaping is applied to maintain correct text formatting.

4️⃣ Removing Duplicate Restaurants: To maintain unique restaurant records, duplicate entries based on restaurant_name are removed, keeping only the first occurrence. This prevents redundancy and ensures each restaurant is represented only once in the dataset.
